##### 2. 테마데이터 수집

```python
theme_dict = {}

for pageNum in range(1, 7):
    url = 'https://finance.naver.com/sise/theme.nhn?page={}'.format(pageNum)
    html = urlopen(url)
    soup = BeautifulSoup(html, "html.parser")
    
    for theme_select in soup.select('#contentarea_left > table.type_1.theme > tr > td.col_type1'):
        try:
            theme_name = theme_select.text
            theme_href = theme_select.select('td > a')[0]['href']
            theme_dict[theme_name] = theme_href
        
        #마지막 테마까지 수집하면, 다음 loop는 index Error
        except:
            break
            
theme_dict
```

```python
theme_stock_dict = {}

for theme, theme_url in theme_dict.items():
    url = 'https://finance.naver.com' + theme_url
    print('url : ', url)
    html = urlopen(url)
    soup = BeautifulSoup(html, "html.parser")
    
    soup.findAll(class_ = 'name_area')
    
    theme_stock_list = [re.sub(pattern = '[*]', repl = '', string = stock.text).strip() for stock in soup.findAll('div', 'name_area')]
    theme_stock_dict[theme] = theme_stock_list
    
theme_stock_dict
```

##### 3. 그룹사

```python
group_stock_dict = {}

for theme, group_url in group_href_dict.items():
    url = 'https://finance.naver.com' + group_url
    print('url : ', url)
    html = urlopen(url)
    soup = BeautifulSoup(html, "html.parser")
    
    soup.findAll(class_ = 'name_area')
    
    group_stock_list = [re.sub(pattern = '[*]', repl = '', string = stock.text).strip() for stock in soup.findAll('div', 'name_area')]
    group_stock_dict[theme] = group_stock_list
    
group_stock_dict
```

```python
kospi_df = pd.read_csv('kospi-stock-code.csv')


#회사가 포함된 섹터에 따라, sector라는 컬럼에 섹터명 입력
for sector, sector_stockList in sector_stock_dict.items():
    kospi_df.loc[kospi_df['회사명'].isin(sector_stockList), 'sector'] = sector


#회사가 포함된 테마에 따라, theme라는 컬럼에 테마명 입력
for theme, theme_stockList in theme_stock_dict.items():
    kospi_df.loc[kospi_df['회사명'].isin(theme_stockList), 'theme'] = theme
    
#회사가 포함된 계열사에 따라, group라는 컬럼에 그룹 입력
for group, group_stockList in group_stock_dict.items():
    kospi_df.loc[kospi_df['회사명'].isin(group_stockList), '계열사'] = group
```

```python
kospi_df = kospi_df.drop(['업종','결산월', '대표자명','지역','상장일'], axis = 1)
```
